TRAINING THE MODEL USING LOGISTIC REGRESSION

```python
input['Duration'] = input['Duration'].str.replace(' min', '').astype(float)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler_data = scaler.fit_transform(input)
scaled_df = pd.DataFrame(scaler_data, columns=input.columns)
scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_df, target, test_size=0.3)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
X_test= np.array(X_test)
```
